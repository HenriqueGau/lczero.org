```python
" Funçao para mostrar matriz"
def m_print(m):
    #para percorrer as linhas da matriz
    for i in m:
        #para cada valor de cada linha
        for valor in i:
            #formatar o vaor para 5d para espaçar
            print(f"{valor:5d}", end="")
        print()
```
